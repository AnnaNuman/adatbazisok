```sql
SELECT SZALL_MOD AS 'SZÁLLÍTÁSI MÓD', COUNT(*) AS 'MENNYISÉG'
FROM Rendeles
GROUP BY SZALL_MOD;
```

```sql
SELECT 
    u.LOGIN, 
    u.NEV AS 'ÜGYFÉL NEVE', 
    COUNT(*) AS 'VÁSÁRLÁSOK SZÁMA'
FROM UGYFEL u
JOIN RENDELES r ON u.LOGIN = r.LOGIN
GROUP BY u.LOGIN, u.NEV
ORDER BY 'VÁSÁRLÁSOK SZÁMA' DESC;
```

```sql
SELECT TERMEKKOD, EGYSEGAR * MENNYISEG AS 'RENDELÉS ÁRA'
FROM Rendeles_tetel
```

```sql
SELECT FIZ_MOD AS 'FIZETÉSI MÓD', COUNT(*) AS 'MENNYISÉG'
FROM Rendeles
GROUP BY FIZ_MOD;
```

```sql
SELECT MEGNEVEZES AS 'TERMÉKEK AMIKBŐL 1-NÉL TÖBBET RENDELTEK', COUNT(*) AS 'Darabszam'
FROM Termek
GROUP BY MEGNEVEZES
HAVING COUNT(*) > 1
```

```sql
SELECT FELVITTE, COUNT(*) AS 'RENDELÉS SZÁM'
FROM Termek
GROUP BY FELVITTE
ORDER BY [RENDELÉS SZÁM] DESC
```

```sql
SELECT 
    CASE 
        WHEN KESZLET < 1000 THEN '0-1000 DB'
        WHEN KESZLET BETWEEN 1000 AND 2000 THEN '1000-2000 DB'
        WHEN KESZLET BETWEEN 2000 AND 5000 THEN '2000-5000 DB'
        ELSE '5000+ DB'
    END AS 'KÉSZLET NAGYSÁGA',
    COUNT(*) AS 'Termekek_szama'
FROM Termek
GROUP BY 
    CASE 
        WHEN KESZLET < 1000 THEN '0-1000 DB'
        WHEN KESZLET BETWEEN 1000 AND 2000 THEN '1000-2000 DB'
        WHEN KESZLET BETWEEN 2000 AND 5000 THEN '2000-5000 DB'
        ELSE '5000+ DB'
    END;
```

```sql
SELECT 
    COUNT(*) AS 'Összes rendelés',
    COUNT(SZALL_CIM) AS 'Van szállítási cím',
    COUNT(*) - COUNT(SZALL_CIM) AS 'Nincs szállítási cím'
FROM Rendeles;
```

```sql
SELECT 
    MIN(LISTAAR) AS 'Minimum ár',
    MAX(LISTAAR) AS 'Maximum ár',
    ROUND(AVG(LISTAAR),2) AS 'Átlag ár'
FROM Termek;
```
